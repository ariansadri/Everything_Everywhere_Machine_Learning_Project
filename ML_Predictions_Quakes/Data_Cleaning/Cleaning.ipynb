```python
# Dependencies
import pandas as pd
import requests
```

```python
earthquake_data = pd.read_csv("clean_earth.csv")
earthquake_data.head()
```

```python
import sys
if sys.version_info[0] < 3:
    from StringIO import StringIO
else:
    from io import StringIO
import pandas as pd
import datetime
```

```python
df = earthquake_data
convert = lambda x: datetime.datetime.fromtimestamp(x / 1e3)
converted_df = df['Time'].apply(convert)
```

```python
df['Time'] = pd.to_datetime(df['Time'], unit='ms')
df.head()
```

```python
df['New_date'] = [d.date() for d in df['Time']]
df['New_time'] = [d.time() for d in df['Time']]
df.head()
```

```python
quake = df[["Magnitude", "New_date", "Place", "State", "City", "Longitude", "Latitude"]]
quake.head()
```

```python
data = quake[["New_date", "Magnitude", "State"]]
data.head()
```

```python
new_data = data[data.State == " CA"]
new_data
```

```python
last_data = new_data[["New_date", "Magnitude"]]
last_data.head()             
```

```python
last_data.rename(columns={'New_date':'Date'}, inplace=True)
last_data.head()
```

```python
last_data.sort_values("Date", ascending=True)
```

```python
newest = last_data.groupby("Date").mean()
newest
```

```python
newest.to_csv('clean_quake_ready.csv')
```
